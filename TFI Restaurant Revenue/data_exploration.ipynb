```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LassoCV
```

```python
train_data = pd.read_csv('train.csv')
test_data = pd.read_csv('test.csv')

train_data['City'] = pd.Categorical(train_data['City'])
train_data['City Group'] = pd.Categorical(train_data['City Group'])
train_data['Type'] = pd.Categorical(train_data['Type'])

train_X = train_data.drop(["Id", "revenue"], axis = 1)
train_y = train_data['revenue']

test_X = test_data.drop(["Id"], axis = 1)
```
